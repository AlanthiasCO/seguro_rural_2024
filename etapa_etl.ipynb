## Filtrando

```python
# no .iloc usamos os índices para selecionar as linhas e colunas que desejamos
# selecionando as primeiras 10 linhas e as primeiras 5 colunas
df_br.iloc[:10,:5]
```

```python
# função loc: seleciona linhas e colunas de acordo com os rótulos   
# ':' = seleciona todas as linhas
# '[]' = seleciona as colunas
df_br.loc[:, ['NR_AREA_TOTAL', 
              'NR_PRODUTIVIDADE_ESTIMADA', 
              'VL_LIMITE_GARANTIA',
              'VL_PREMIO_LIQUIDO', 
              'VL_SUBVENCAO_FEDERAL', 
              'NR_APOLICE']]
```

```python
# Obtendo a quantidade de municípios presentes no dataframe
df_br.loc[df_br['NM_MUNICIPIO_PROPRIEDADE'] == 'Varginha']
```
